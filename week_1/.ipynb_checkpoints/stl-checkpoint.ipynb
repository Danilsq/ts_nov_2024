### Задание 1 - реализовать "наивную" имплементацию stl-разложения:
    Ряд - stl_example.csv

    0. Убрать гетероскедастичность логарифмированием.
    1. Апроксимировать ряд линейным трендом.
    2. Вычесть линейный тренд из ряда.
    3. Найти период сезонности получившегося ряда по коррелограме.
    4. Получить сезонность ряда без тренда одним из двух способов: 
       а) при помощи медианного фильтра с окном равным период/к, к подобрать, обычно хорошо подходит k={2, 3, 4}. 
       (Чтобы фильтр считался по центру скользящего окна, используйте опцию ts.rolling(k, center=True).median())
       б) продифференцировав его с лагом, равным найденному периоду сезонности и вычтя
       получившийся ряд из исходного
       в)* попробуйте подумать, какие есть плюсы и минусы у каждого из способов?
    5. Вычесть тренд и сезонность, получить остатки.
    6. Проверить остатки на стационарность.

   
detect_ts должна возвращать tuple из: (тренд, сезонность, остатки)

```python
dataset = Dataset('../data/dataset/')
```

```python
ts = dataset['stl_example.csv']
```

```python
# Для задания 1 из homework_1:
# У исходного ряда присутствует линейный тренд и сезонность.
# Исходя из графиков ACF и PACF видно, что у ряда есть 6 часовая сезонность или 3 часовая сезонность с
# движением в разные стороны f(x)=(-1)^(T/3)*f(x+T), T=3. А также часовая сезонность в виду близких значений
# в окрестности x.
# f(x)=-f(x+3) можно представить через нечетную функцию с периодом 3. например, sin(pi/3 * x)
```

```python
ts.plot(figsize=(10,5))
```

```python
tsaplots.plot_acf(ts, lags=50);
tsaplots.plot_pacf(ts, lags=50);
```

```python
def transform_series(ts: pd.Series, func):
    """
    Преобразует ряд
    """
    return ts.apply(func)

def extract_trend(ts: pd.Series):
    """
    Извлекает линейный тренд из временного ряда
    """
    k, b = np.polyfit(range(len(ts)), ts, 1)
    trend = pd.Series(range(len(ts)) * k + b, index = ts.index)
    
    return trend, k, b

def extract_seasonality(ts_detrended: pd.Series, period=None):
    """
    Извлекает сезонную компоненту
    """
    season = ts_detrended.rolling(int(period/2), center = True).median()
    
    return season

def extract_seasonality_diff(ts_detrended: pd.Series, period=None):
    """
    Извлекает сезонную компоненту при помощи дифференцирования
    """
    season = (ts_detrended-ts_detrended.diff(period))
    
    return season

def detect_ts(ts):
    ts_trans = transform_series(ts, np.log)
    trend = extract_trend(ts_trans)[0]
    ts_detrended = ts_trans-trend
    season = extract_seasonality_diff(ts_detrended, 6)
    resid = ts_detrended - season

    return (trend, season, resid)
```

```python
# Убрать гетероскедастичность логарифмированием.
ts_transorm = transform_series(ts, np.log)
```

```python
ts_transorm.plot(figsize=(10,5))
```

```python
# Апроксимировать ряд линейным трендом.
trend, k, b = extract_trend(ts_transorm)
```

```python
ts_transorm.plot(figsize=(10,5))
trend.plot()
```

```python
# Вычесть линейный тренд из ряда.
ts_transorm_noTrend = ts_transorm - trend
```

```python
ts_transorm_noTrend.plot(figsize=(10,5))
```

```python
tsaplots.plot_acf(ts_transorm_noTrend, lags=50);
tsaplots.plot_pacf(ts_transorm_noTrend, lags=50);
```

```python
# Получить сезонность ряда без тренда одним из двух способов: 
# а) при помощи медианного фильтра с окном равным период/к, к подобрать, обычно хорошо подходит k={2, 3, 4}. 
# (Чтобы фильтр считался по центру скользящего окна, используйте опцию ts.rolling(k, center=True).median())
season_1 = extract_seasonality(ts_transorm_noTrend, 6)
ts_transorm_noTrend.plot(figsize=(10,5))
season_1.plot()
```

```python
# Получить сезонность ряда без тренда одним из двух способов: 
# б) продифференцировав его с лагом, равным найденному периоду сезонности и вычтя
# получившийся ряд из исходного
season_2 = extract_seasonality_diff(ts_transorm_noTrend, 6)
ts_transorm_noTrend.plot(figsize=(10,5))
season_2.plot()
```

```python
# Получить сезонность ряда без тренда одним из двух способов: 
# в)* попробуйте подумать, какие есть плюсы и минусы у каждого из способов?
# 1 метод не реагирует на шоки, 2 метод, наоборот, может предсказывать шок, когда его нет
# идеальным приближением будет функция от sin(pi/3 * x), но это не считается, так как ряд из нее и получался ;)
ts_transorm_noTrend.plot(figsize=(10,5))
ts_transorm_noTrend.apply(lambda x: np.sin(np.pi * x / 3)).plot()
```

```python
# Вычесть тренд и сезонность, получить остатки.
resid = ts_transorm_noTrend - season_2
```

```python
resid.plot(figsize = (10,5))
```

```python
# Проверить остатки на стационарность.
# остатки стационарны
print(f'P-value теста Дики-Фуллера: {adfuller(resid.dropna())[1]:.4f}')
```

```python
trend, season, resid = detect_ts(ts)
fig, ax = plt.subplots(3, figsize = (10,5))
ax[0].plot(trend)
ax[1].plot(season)
ax[2].plot(resid)

ax[0].set_title('Trend')
ax[1].set_title('Season')
ax[2].set_title('Resid')

plt.tight_layout()
plt.show()
```

### Задание 2 - найти аномалии во временном ряде при помощи получившихся остатков 

1. Расчитать стандартное отклонение остатков std
2. Получить порог на остатки по формуле `threshold = k * std`, k обычно берется от 2 до 3.
3. Найти аномалии, как точки ряда, абсолютные значения которых превышают найденный порог

```python
resid.std()
```

```python
treshold = resid.std() * 3
```

```python
outliers = resid[abs(resid) > treshold]
```

```python
outliers
```

```python
resid.plot(figsize = (10,5), label = 'resid')
plt.hlines(treshold, resid.index.min(), resid.index.max(), colors = 'black', linestyles = '--', label = 'treshold')
plt.hlines(-treshold, resid.index.min(), resid.index.max(), colors = 'black', linestyles = '--')
plt.scatter(outliers.index, outliers.values, color = 'red', s = 10, label = 'outliers')
plt.legend()
plt.show()
```

### Задание 3 - сделать прогноз ряда на 6 периодов вперед (36 точек)

1. Экстраполируйте линейный тренд.
2. Сделайте рекурсивный прогноз сезонной компоненты по формуле y(t) = y(t-6)
3. Остатки по-хорошему должны моделироваться моделью arma, но в нашем случае сделайте просто прогноз средним значением. (Т.к. в нашем случае это 0, остатки можно вообще проигнорировать)
4. Сложите получившиеся компоненты и получите финальный прогноз
5. profit!

```python
periods = 36
dt_forward = pd.date_range(start=ts.index[-1] + pd.Timedelta(hours=1), periods=periods, freq = '1h')
```

```python
# Экстраполируйте линейный тренд.
trend_predict = range(len(ts), len(ts)+periods)*k + b
trend_predict = pd.Series(trend_predict, index = dt_forward)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(trend, label='Original Series')
plt.plot(trend_predict, label='Predicted Series')
plt.legend()
plt.show()
```

```python
# Сделайте рекурсивный прогноз сезонной компоненты по формуле y(t) = y(t-6)
season_predict = pd.Series(index = dt_forward)
season_predict = pd.concat([season[-6:], season_predict])
for i in range(6, 6+periods):
    season_predict.iloc[i] = season_predict.iloc[i-6]
season_predict = season_predict.iloc[6:]
```

```python
plt.figure(figsize=(10, 5))
plt.plot(season, label='Original Series')
plt.plot(season_predict, label='Predicted Series')
plt.legend()
plt.show()
```

```python
# Остатки по-хорошему должны моделироваться моделью arma, но в нашем случае сделайте просто прогноз средним значением. 
# (Т.к. в нашем случае это 0, остатки можно вообще проигнорировать)
np.random.seed(seed = 44)
resid_predict = np.random.normal(loc = resid.mean(), scale = resid.std(), size = periods)
resid_predict = pd.Series(resid_predict, dt_forward)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(resid, label='Original Series')
plt.plot(resid_predict, label='Predicted Series')
plt.legend()
plt.show()
```

```python
# Сложите получившиеся компоненты и получите финальный прогно.
ts_predict = trend_predict + season_predict + resid_predict
```

```python
plt.figure(figsize=(10, 5))
plt.plot(ts_transorm, label='Original Series')
plt.plot(ts_predict, label='Predicted Series')
plt.legend()
plt.show()
```
