```python
import pandas as pd
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from dataset import Dataset
from statsmodels.tsa.statespace import sarimax
from joblib import Parallel, delayed
import itertools
from sklearn.metrics import mean_squared_error
import numpy as np

%matplotlib inline
import matplotlib.pyplot as plt

import warnings
warnings.simplefilter('ignore')
```

```python
dataset = Dataset('../data/dataset/', header = None)
```

```python
dj = dataset['dow_jones_0.csv']
```

```python
dj.plot(figsize = (10,5))
```

```python
adfuller(dj_diff.dropna())
```

```python
plot_acf(dj_diff.dropna(), lags = 50);
plot_pacf(dj_diff.dropna(), lags = 50);
```

```python
split_share = 0.7

y_train = dj[:int(split_share*len(dj))]
y_test = dj[int(split_share*len(dj)):]
```

```python
params = {
    'order': {
        'p': range(0, 3),
        'd': range(0, 3),
        'q': range(0, 3)
    },
    'seasonal_order': {
        'P': range(0, 3),
        'D': [0],
        'Q': range(0, 3),
        'm': [0,12]
    }
}
```

```python
def grid_search(model_class, y_train, y_test, params):
    
    order = params['order']
    seasonal_order = params['seasonal_order']
    param_combinations = list(itertools.product(
        order['p'], order['d'], order['q'],
        seasonal_order['P'], seasonal_order['D'], seasonal_order['Q'], seasonal_order['m'])
                             )
    
    def model_enumeration(combination):
        p, d, q, P, D, Q, m = combination
        order = (p, d, q)
        seasonal_order = (P, D, Q, m)
        try:
            model = model_class(y_train, order = order, seasonal_order = seasonal_order)
            model_fit = model.fit(disp = False)
            y_hat = model_fit.predict(start=len(y_train), end=len(y_train)+len(y_test)-1, dynamic = True)
            rmse = np.sqrt(mean_squared_error(y_test, y_hat))
            return {'params': {
                'order':dict(zip(('p','d','q'), order)),
                'seasonal_order': dict(zip(('P','D','Q', 'm'), seasonal_order))
            }, 'rmse': rmse}
        
        except:
            return {'status': 'BAD_FIT'}
            
    results = Parallel(n_jobs=-1)(delayed(model_enumeration)(combo) for combo in param_combinations)
    results_sorted = sorted(results, key=lambda result: result.get('rmse', float('inf')))
    best_params = results_sorted[0]['params']

    return best_params
```

```python
best_params = grid_search(sarimax.SARIMAX, y_train, y_test, params)
```

```python
best_params
```

```python
best_model = sarimax.SARIMAX(y_train, 
                             order = tuple(best_params['order'].values()), 
                             seasonal_order = tuple(best_params['seasonal_order'].values()))
best_model = best_model.fit(disp = False)
```

```python
y_hat = best_model.predict(start=len(y_train), end=len(y_train)+len(y_test)-1, dynamic = True) 
```

```python
mse = np.sqrt(mean_squared_error(y_test, y_hat))
print(f'RMSE: {mse:.4f}')
```

```python
plt.figure(figsize=(12, 6))
plt.plot(y_train.index, y_train, label='Train', color='blue')
plt.plot(y_test.index, y_test, label='Test', color='orange')
plt.plot(y_test.index, y_hat, label='Forecast', color='red')
plt.title('Dow Jones over Time with SARIMA Forecast')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
